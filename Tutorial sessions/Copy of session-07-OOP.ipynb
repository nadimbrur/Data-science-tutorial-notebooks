```python
# Two rules for create variable in the class i) delare inside the constructor ii) using 'self'

# Golden rule of OOP: class element (attribute, method) only acessed by the object.
class Atm:
  # constructor(special function)->superpower -> (run automatically whenever we create a object)
  def __init__(self):

    #These are attribute
    print(id(self))
    self.pin = ''
    self.balance = 0
    # self.menu()

  # These are function or method
  def menu(self):
    user_input = input("""
    Hi how can I help you?
    1. Press 1 to create pin
    2. Press 2 to change pin
    3. Press 3 to check balance
    4. Press 4 to withdraw
    5. Anything else to exit
    """)

    if user_input == '1':
      self.create_pin()
    elif user_input == '2':
      self.change_pin()
    elif user_input == '3':
      self.check_balance()
    elif user_input == '4':
      self.withdraw()
    else:
      exit()

  def create_pin(self):
    user_pin = input('enter your pin')
    self.pin = user_pin

    user_balance = int(input('enter balance'))
    self.balance = user_balance

    print('pin created successfully')
    self.menu()

  def change_pin():
    old_pin = input('enter old pin')

    if old_pin == self.pin:
      # let him change the pin
      new_pin = input('enter new pin')
      self.pin = new_pin
      print('pin change successful')
      self.menu()
    else:
      print('nai karne de sakta re baba')
      self.menu()

  def check_balance(self):
    user_pin = input('enter your pin')
    if user_pin == self.pin:
      print('your balance is ',self.balance)
    else:
      print('chal nikal yahan se')

  def withdraw(self):
    user_pin = input('enter the pin')
    if user_pin == self.pin:
      # allow to withdraw
      amount = int(input('enter the amount'))
      if amount <= self.balance:
        self.balance = self.balance - amount
        print('withdrawl successful.balance is',self.balance)
      else:
        print('abe garib')
    else:
      print('sale chor')
    self.menu()

  def xyz(self):
    print(id(self))
```

### Magic method
i) Constrcutor

```python
# i) Constructor: Is a method, special method have superpower (run automatically when object created).
# __init__()

# Advantages:
# i) Initialize
# ii) Some things that we don't want to give the access to the user that can be done by using __init__(self) like 'Internet auto connected',
# connected with backend (configutation related).
```

```python

class Fraction:
  # parameterized constructor
  # Self: Current object
  def __init__(self,x,y):
    self.num = x
    self.den = y


  # Magic method 2) __str__ for print
  def __str__(self):
    # For print
    return '{}/{}'.format(self.num,self.den)

  # Magic method 2) __add__ for '+'
  def __add__(self,other):
    print(id(self),id(other))
    # Algorithm for add two object
    new_num = self.num*other.den + other.num*self.den
    new_den = self.den*other.den

    return '{}/{}'.format(new_num,new_den)

  # Magic method 2) __sub__ for '-'
  def __sub__(self,other):
    # Algorithm for sub two object
    new_num = self.num*other.den - other.num*self.den
    new_den = self.den*other.den

    return '{}/{}'.format(new_num,new_den)

  # Magic method 2) __mul__ for '*'
  def __mul__(self,other):
    # Algorithm for mul two object
    new_num = self.num*other.num
    new_den = self.den*other.den

    return '{}/{}'.format(new_num,new_den)

  # Magic method 2) __truediv__ for '/'
  def __truediv__(self,other):
    new_num = self.num*other.den
    new_den = self.den*other.num

    # Algorithm for division two object
    return '{}/{}'.format(new_num,new_den)

  def convert_to_decimal(self):
    return self.num/self.den
```

```python
fr1 = Fraction(3,4)
fr2 = Fraction(1,2)
```

```python
0+fr2
# 3/4
```

```python
print(fr1 + fr2)
print(fr1 - fr2)
print(fr1 * fr2)
print(fr1 / fr2)
```
